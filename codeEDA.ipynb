# **Feature Selection: Weight of evidence(WOE) & information value(IV)**

```python
# WoE & IV for categorical variables
#WoE=ln(Good loan%/Bad loan%)
def WoE_categorical(df,feature):
    df[feature+'_WoE']=df[feature].map(
        np.log(
            ((df.groupby(feature).TARGET.count()-df.groupby(feature).TARGET.sum())/(df.TARGET.count()-df.TARGET.sum()))/(
            df.groupby(feature).TARGET.sum()/df.TARGET.sum()
            )
        )
    )
    df[feature+'_WoE']=df[feature+'_WoE'].to_list()
    df.replace([np.inf, -np.inf], 0, inplace=True)
    return df

#IV=WoE*(good loan%-bad loan%)

def IV_categorical(df,features):
    IV_table={}
    for item in features:
        #print('WoE:',df.groupby(item)[item+'_WoE'].mean())
        #print('item:',item)
        All_Good=df.TARGET.count()-df.TARGET.sum()
        All_Bad=df.TARGET.sum()
        IV_table[item]=(df.groupby(item)[item+'_WoE'].mean()*(
            ((df.groupby(item).TARGET.count()-df.groupby(item).TARGET.sum())/All_Good-df.groupby(item).TARGET.sum()/All_Bad)
        )).sum().round(8)
    IV_table=pd.DataFrame.from_dict(IV_table,orient='index',columns=['IV']).reset_index().rename(columns={'index':'Features'}).sort_values(by='IV')
    return IV_table
```

```python
Cat_col=Train_application.columns[Train_application.dtypes=='object']
Cat_col
```

```python
df=Train_application[Train_application.columns[~Train_application.columns.isin(Cat_col)]]
all_cat_col=df.columns[df.nunique()<=5].to_list()
all_cat_col.remove('TARGET')
all_cat_col=all_cat_col+Cat_col.to_list()
```

```python
#Check WoE & IV of categorical features
Train_application_WoE=Train_application.copy()
for item in all_cat_col:
    Train_application_WoE=WoE_categorical(Train_application_WoE,item)

IV_table=IV_categorical(Train_application_WoE,all_cat_col)
```

When information value is smaller than 0.02, the feature is useless for prediction, we will remove such feature from our dataset:

- IV < 0.02: not useful for prediction
- 0.02 ≤ IV < 0.1: weak predictor
- 0.1 ≤ IV < 0.3: moderate predictor
- 0.3 ≤ IV < 0.5: strong predictor
- IV ≥ 0.5: suspicious predictor
